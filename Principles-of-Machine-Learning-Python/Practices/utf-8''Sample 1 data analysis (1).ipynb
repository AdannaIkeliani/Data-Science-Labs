```python
def encode_string(data_feature):
    enc = preprocessing.LabelEncoder()
    enc.fit(data_feature)
    enc_data_feature = enc.transform(data_feature)
    ohe = preprocessing.OneHotEncoder()
    encoded = ohe.fit(enc_data_feature.reshape(-1,1))
    return encoded.transform(enc_data_feature.reshape(-1,1)).toarray()

category_column = ['Mood','Weather', 'HaveTransport','Traffic Jam','SleepingPartner','CometoWork']

for col in category_column:
    temp = encode_string(data[col])
    features = np.concatenate([features,temp], axis = 1)
    
print(features.shape)
print(features[:2, :])
    
```

```python
#adding numeric features

features= np.concatenate([features, np.array(data[['Task']])], axis = 1)
features[:2,:]

```

```python
features.shape
```

```python
#splitting the data
# from sklearn.model_selection import train_test_split
# x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.25, random_state=0)
import numpy.random as nr
import sklearn.model_selection as ms
nr.seed(3725)
labels = np.array(data['CometoWork'])
indx = range(features.shape[0])
indx = ms.train_test_split(indx, test_size = 17)
x_train = features[indx[0],:]
y_train = np.ravel(labels[indx[0]])
x_test = features[indx[1],:]
y_test = np.ravel(labels[indx[1]])
```

```python
#rescale numeric features
scaler = preprocessing.StandardScaler().fit(x_train[:,22:])
x_train[:,22:] = scaler.transform(x_train[:,22:])
x_test[:,22:] = scaler.transform(x_test[:,22:])
print(x_train.shape)
x_train[:5,:]
```
